```python
import sys
data = sys.stdin.read().strip().split()
n = int(data[0])
photos = 0

for i in range(n):
	y = int(data[2*i + 1])
    p = float(data[2*i + 2])
    photos.append((p,y))
photos.sort(reverse=True)
tp = 0
fp = 0
precisions = []
thresholds = []
for p,y in photos:
    if y==1:
        tp+=1
    else:
        fp+=1
    precision = tp/(tp+fp)
    precisions.append(precision)
    thresholds.append(p)

for t, pr in zip(thresholds, precisions):
    print(f"{t:.6f} {pr:.6f}")
```
